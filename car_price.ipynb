```python
import seaborn as sns
import matplotlib.pyplot as plt


def get_count_plots(df, cols_list):
    for col in cols_list:
        plt.figure(figsize=(3,3))
        sns.countplot(data=df, x=col)
        plt.title(f'Value Counts of {col}')
        plt.xticks(rotation=45)
        plt.show()
```

```python
df['CarName']
```

```python
df['brand']= df['CarName'].apply(lambda x : x.split(' ')[0])
```

```python
df['brand'].unique()
```

```python
df['brand']= df['brand'].replace({'maxda': 'mazda', 'nissan':'Nissan', 'porcshce':'porsche', 'vokswagen':'volkswagen', 'vw':'volkswagen', 'toyouta':'toyota'})
```

```python
df['brand'].unique()
```

```python
df['model'] = df['CarName'].apply(lambda x: ' '.join(x.split(' ')[1:]))
```

```python
df['model'].unique()
```

```python
df.columns
```

```python
categorical_columns = ['brand','fueltype','aspiration','doornumber','carbody','enginelocation','enginetype','cylindernumber','fuelsystem']
```

```python
for i in categorical_columns:
    print(f"category in {i} is : {df[i].unique()}")
```

```python
df['fueltype']= df['fueltype'].replace({'gas': 0, 'diesel':1}) 
```

```python
df['aspiration']= df['aspiration'].replace({'std': 0, 'turbo':1}) 
```

```python
df['doornumber']= df['doornumber'].replace({'two': 2, 'four':4}) 
```

```python
df['enginelocation']= df['enginelocation'].replace({'front': 0, 'rear':1}) 
```

```python
df['cylindernumber']= df['cylindernumber'].replace({'four': 4, 'six':6, 'five':5, 'three':3, 'twelve':12, 'two':2, 'eight':8}) 
```

```python
one_hot = pd.get_dummies(df['drivewheel'], dtype=int, drop_first=True)
# Drop column B as it is now encoded
df = df.drop('drivewheel',axis = 1)
# # Join the encoded df
df = df.join(one_hot)
```

```python
df
```

```python
label_columns = ['brand','carbody','fuelsystem','enginetype']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
for columns in label_columns:
    df[columns] = enc.fit_transform(df[columns])
```
